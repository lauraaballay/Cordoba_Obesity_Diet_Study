```python
def get_country(location):
    country=str(location).lower()
    if country.find('argentina')>=0:
        return 'argentina'
    elif country.find('arg')>=0:
        return 'argentina'
    elif get_province(location) != '':
        return 'argentina'
    else:
        return ''

def get_province(location):
    province=str(location).lower()

    if province.find('buenos aires')>=0:
        return 'buenos aires'
    if province.find('zárate')>=0:
        return 'buenos aires'
    elif province.find('catamarca')>=0:
        return 'catamarca'
    elif province.find('chaco')>=0:
        return 'chaco'
    elif province.find('chubut')>=0:
        return 'chubut'
    elif province.find('corrientes')>=0:
        return 'corrientes'
    elif province.find('cordoba')>=0:
        return 'cordoba'
    elif province.find('córdoba')>=0:
        return 'cordoba'
    elif province.find('entre ríos')>=0:
        return 'entre rios'
    elif province.find('entre rios')>=0:
        return 'entre rios'
    elif province.find('formosa')>=0:
        return 'formosa'
    elif province.find('jujuy')>=0:
        return 'jujuy'
    elif province.find('la pampa')>=0:
        return 'la pampa'
    elif province.find('la rioja')>=0:
        return 'la rioja'
    elif province.find('mendoza')>=0:
        return 'mendoza'
    elif province.find('misiones')>=0:
        return 'misiones'
    elif province.find('neuquén')>=0:
        return 'neuquen'
    elif province.find('neuquen')>=0:
        return 'neuquen'
    elif province.find('río negro')>=0:
        return 'rio negro'
    elif province.find('rio negro')>=0:
        return 'rio negro'
    elif province.find('salta')>=0:
        return 'salta'
    elif province.find('san juan')>=0:
        return 'san juan'
    elif province.find('san luis')>=0:
        return 'san luis'
    elif province.find('santa cruz')>=0:
        return 'santa cruz'
    elif province.find('santa fé')>=0:
        return 'santa fe'
    elif province.find('santa fe')>=0:
        return 'santa fe'
    elif province.find('santiago del estero')>=0:
        return 'santiago del estero'
    elif province.find('tierra del fuego')>=0:
        return 'tierra del fuego'
    elif province.find('tucumán')>=0:
        return 'tucuman'
    elif province.find('tucuman')>=0:
        return 'tucuman'
    else:
        return ''
```

```python
dataset['province'] = dataset.apply(lambda x: get_province(x['location']),axis=1)
```

```python
dataset['country'] = dataset.apply(lambda x: get_country(str(x['location'])),axis=1)
```
